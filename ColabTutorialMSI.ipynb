Poniżej film instruktażowy jak ściągnąć kaggle.json

```python
from IPython.display import HTML
from base64 import b64encode
 
def show_video(video_path, video_width = 600):
   
  video_file = open(video_path, "r+b").read()
 
  video_url = f"data:video/mp4;base64,{b64encode(video_file).decode()}"
  return HTML(f"""<video width={video_width} controls><source src="{video_url}"></video>""")
 
show_video("kaggle_json_download.mp4")
```

```python
number_input = 20 #@param {type:"number"}
number_slider = 0.3 #@param {type:"slider", min:-1, max:1, step:0.1}

integer_input = 5 #@param {type:"integer"}
integer_slider = 12 #@param {type:"slider", min:0, max:100, step:1}

print(number_input)
print(number_slider)

print(integer_input)
print(integer_slider)
```
